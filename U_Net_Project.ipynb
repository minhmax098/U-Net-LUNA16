```python
import os, glob
import SimpleITK as sitk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

DATA_DIR = "/content/drive/MyDrive/LUNA16"
```

```python
# read file annotation
annotation = pd.read_csv(os.path.join(DATA_DIR, "annotations.csv"))
print("Total number of nodules: ", len(annotation))
annotation.head()
```

```python
# Get all list file .mhd in subset0 and subset1
mhd_files = []
for subset in ["subset0", "subset1"]:
  mhd_files += glob.glob(os.path.join(DATA_DIR, subset, subset, "*.mhd"))

print("Number of CT in first 2 subsets:", len(mhd_files))

# map seriesuid -> path .mhd
series_to_path = {
    os.path.splitext(os.path.basename(p))[0]: p
    for p in mhd_files
}

print ("Example of first 3 series: ")
list(series_to_path.items())[:3]
```

```python
annotation_01 = annotation[annotation["seriesuid"].isin(series_to_path.keys())]

print("Number of nodules subset0, subset1: ", len(annotation_01))
print("Number of CT have nodules in subset 0, subset1: ", annotation_01["seriesuid"].nunique())
annotation_01.head()
```

```python
# get 1 line annotation in subset0-subset1
row = annotation_01.iloc[0]
seriesuid = row["seriesuid"]
print("Series choose: ", seriesuid)

ct_path = series_to_path[seriesuid]
print("Path CT: ", ct_path)
```

```python
# read volume
image = sitk.ReadImage(ct_path)
volume = sitk.GetArrayFromImage(image)

print("Volume shape (x, y, z): ", volume.shape)
print("Volume shape (z, y, x): ", volume.shape)
print("Volume shape (y, z, x): ", volume.shape)
print("Spacing (x, y, z): ", image.GetSpacing())
print("Origin: ", image.GetOrigin())
```

```python
# Create mask 3D for all nodules in series
# all nodules in the same series
nodes = annotation_01[annotation_01["seriesuid"] == seriesuid]
print("Number of nodules in this scan: ", len(nodes))

mask = np.zeros(volume.shape, dtype=np.uint8)

spacing = np.array(image.GetSpacing())
sx, sy, sz = spacing[0], spacing[1], spacing[2]

# create voxel coordinate (z, y, x)
zz, yy, xx = np.ogrid[:volume.shape[0], :volume.shape[1], :volume.shape[2]]

for _, n in nodes.iterrows():
  # coordinate world (mm)
  world = (float(n.coordX), float(n.coordY), float(n.coordZ))
  # tranfer index voxel (x, y, z) using SimpleITK
  cx, cy, cz = image.TransformPhysicalPointToIndex(world)
  # convert to (z, y, x) order for numpy
  cz, cy, cx = int(cz), int(cy), int(cx)

  radius_mm = float(n.diameter_mm) / 2.0

  # distance in mm
  distance2 = ((xx - cx) * sx) ** 2 + ((yy - cy) * sy) ** 2 + ((zz - cz) * sz) ** 2
  mask[distance2 <= radius_mm ** 2] = 1
```

```python
# Take a slice near the first nodes
first_nodule = nodes.iloc[0]
cx, cy, cz = image.TransformPhysicalPointToIndex(
    (float(first_nodule.coordX), float(first_nodule.coordY), float(first_nodule.coordZ))
)

# z-index
slice_idx = int(cz)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(volume[slice_idx], cmap="gray")
plt.title(f"CT slice {slice_idx}")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(volume[slice_idx], cmap="gray")

# overlay mask
plt.imshow(mask[slice_idx], alpha=0.3)
plt.title("CT + nodule mask")
plt.axis("off")

plt.tight_layout()
plt.show()
```
